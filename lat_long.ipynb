```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/Users/obus_/OneDrive/Desktop/IBM DS/toronto.csv')
df.head()
```

```python
df_latlng = pd.DataFrame(columns = ['Postal Code', 'Latitude', 'Longitude'])
df_latlng['Postal Code'] = df['Postal Code']
df_latlng
```

```python
import pgeocode
nomi = pgeocode.Nominatim('ca')
n = df_latlng['Postal Code'].size

for i in range(0,n):
    df_latlng['Latitude'][i] = nomi.query_postal_code(df['Postal Code'][i]).latitude
    df_latlng['Longitude'][i] = nomi.query_postal_code(df['Postal Code'][i]).longitude
    
df_latlng.head()
```

```python
df['Latitude'] = df_latlng['Latitude']
df['Longitude'] = df_latlng['Longitude']
```
